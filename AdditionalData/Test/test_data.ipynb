Read in data - FFIEC CDR Call Subset of Schedules 2019(1 of 2).txt

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mlt

pd.set_option('mode.chained_assignment', None)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

data = pd.read_csv("FFIEC CDR Call Subset of Schedules 2019(1 of 2).txt",sep='\t', chunksize=100000, low_memory=False)

df_raw = pd.DataFrame()

for chunk in data:
    df_raw=df_raw.append(chunk)
```
