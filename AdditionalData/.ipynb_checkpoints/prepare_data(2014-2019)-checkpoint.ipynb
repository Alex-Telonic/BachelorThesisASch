```python
def choose_securities(gr):
    if gr.iloc[1].year==2019:
        print(True)
        gr.securities = gr.securities2019
    return gr

df_2014_2019 = df_2014_2019.groupby(["year"]).apply(choose_securities)
```

```python
def compute_loansnet(gr):
    if gr.year.isin(range(2014,2017)).any():
        gr.loansnet.fillna(0, inplace=True)
        gr["loansnet"] = gr["loansnet"] + gr["unearnedincome"]
        print(gr[gr.unearnedincome.notnull()][["loansnet", "unearnedincome"]],)
    return gr

df_2014_2019 = df_2014_2019.groupby(["year"]).apply(compute_loansnet)
```
